# 전처리 과정
1. 각 언어별로 문자가 20개씩 들어있는 삼중 폴더 형식을 images폴더에 하나로 저장
2. 즉, 기존에 miniimagenet과 동일하게 구성
3. label은 png파일의 앞에 4글자에 'omni'접두사를 붙여 사용('omni0709')

## 폴더 구성

```python
folder_list = os.listdir('images/')
folder_list
```

```python
folder_dict = {}
for folder in folder_list:
    folder_dict[folder] = os.listdir(os.path.join('images', folder))
    
    
folder_dict
```

### Main dict -> label_naming

```python
label_naming = {}
base_path = 'images/'

for lang in folder_dict.keys():
    items = folder_dict[lang]
    for v in items:
        label_naming[os.path.join(base_path, lang, v)] = os.listdir(os.path.join(base_path, lang, v))
```

```python
label_naming
```

## tag labeling

```python
prefix = 'omni'
labels = {'label':[], 'tag':[]}

for key,values in label_naming.items():
    tag = '/'.join(key.split('/')[1:]) # ex) Korean/character01
    label = prefix+values[0][:4] # ex) omni0709
    
    labels['label'].append(label)
    labels['tag'].append(tag)
    
labels
```

```python
labels_df = pd.DataFrame(labels, columns=['label', 'tag'])

labels_df
```

```python
labels_df.to_csv('label.csv', index=False)
```

## image moving(renaming)

```python
import shutil

base_path = 'images'
prefix = 'omni'
for key,values in label_naming.items():
    for v in values:
        old_path = os.path.join(key, v)
        new_path = os.path.join(base_path, prefix+v)
        
        new_file_path = shutil.move(old_path, new_path)
        print(new_file_path)
    print()
```

```python
images_list = os.listdir('images/')
len(images_list)
```

```python
for file in images_list:
    if file[-3:] != 'png':
        print(file)
        shutil.rmtree(os.path.join(base_path, file))
```

```python
images_list = os.listdir('images/')
len(images_list)
```

```python
total_dict = {'file_path':[], 'label':[]}

for file in images_list:
    label = file[:8]
    total_dict['file_path'].append(file)
    total_dict['label'].append(label)
    
total_dict
```

```python
total_csv = pd.DataFrame(total_dict, columns=['file_path', 'label'])
total_csv
```

```python
total_csv.to_csv('omniglot.csv', index=False)
```

## train.csv & valid.csv & test.csv 생성

```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
file_list = os.listdir('images/')
len(file_list)
```

```python
labels = pd.read_csv('label.csv')
label = labels['label'].tolist()
np.random.shuffle(label)

len(label)
```

```python
train_label = label[:1200]; label = label[1200:]; np.random.shuffle(label)
valid_label = label[:120]; label = label[120:]; np.random.shuffle(label)
test_label = label

print(len(train_label))
print(len(valid_label))
print(len(test_label))
```

```python
train_csv = {'file_path': [], 'label':[]}
valid_csv = {'file_path': [], 'label':[]}
test_csv = {'file_path': [], 'label':[]}

for i, file in enumerate(file_list):
    l = file[:8]
    if l in train_label:
        train_csv['file_path'].append(file)
        train_csv['label'].append(l)
    elif l in valid_label:
        valid_csv['file_path'].append(file)
        valid_csv['label'].append(l)
    else:
        test_csv['file_path'].append(file)
        test_csv['label'].append(l)
```

```python
print(len(train_csv['file_path']))
print(len(valid_csv['file_path']))
print(len(test_csv['file_path']))
```

```python
train = pd.DataFrame(train_csv, columns=['file_path', 'label'])
valid = pd.DataFrame(valid_csv, columns=['file_path', 'label'])
test = pd.DataFrame(test_csv, columns=['file_path', 'label'])
```

```python
train.to_csv('train.csv', index=False)
valid.to_csv('valid.csv', index=False)
test.to_csv('test.csv', index=False)
```
